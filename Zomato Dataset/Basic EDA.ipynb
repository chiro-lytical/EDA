```python
# Lets try to see Country Code
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# Let's Combine These 2 DataFrames
final_df=pd.merge(df, df_country, on='Country Code', how='left')

# on = tells on which feature you want to merge
# how = tells on which one you want to foucs, type of joint
```

```python
final_df.head(3)
```

```python
final_df.Country.value_counts()
# Tells how many countries are there and how many counts related to them
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# PIE CHART - Top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observation:
Zomato maximum records or transactions are from India, followed by USA and UK

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Relationship Between Rating Color and Aggregate Rating:
Goes from Red to Dark Green (Poor to Excellent)
1. 4.5 to 4.9---> Excellent
2. 4.0 to 4.4---> Very Good
3. 3.5 to 3.9---> Good
4. 2.5 to 3.4---> Average
5. 2.0 to 2.4---> Poor
- Maximum Poeple who use the app have not given ratings
- For those who have, let's plot the graph

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```

Observation:
1. Not Rated count is very high
2. Maximum ratings are between 2.5 to 3.4


```python
sns.countplot(x="Rating color", data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```
